```python
import pandas as pd
from pandas import DataFrame, Series

df = pd.read_csv("P12_input001.csv")
```

```python
type(df)
```

```python
df= df.rename(columns= {'0':'t', '0.1':'P_t'})
df
```

```python
import warnings
import itertools
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df['P_t'].plot(figsize=(12,8), title='P_t history', fontsize=14)
```

```python
df['P_t'] = np.log(df['P_t']) 
df['P_t'].plot(figsize=(12,8), title='P_t history(log scale)', fontsize=14)
```

```python
decomposition = seasonal_decompose(df.P_t, freq=63) # seasonal을 통해 시계열 데이터의 주기성을 파악하고자 함

fig = plt.figure()
fig = decomposition.plot()
fig.set_size_inches(12,16) 
```
